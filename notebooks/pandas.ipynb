## Pandas Tutorial

```python
import pandas as pd
```

```python
file = pd.read_csv("https://archive.ics.uci.edu/static/public/172/data.csv")
file.head()
```

```python
file.info()
```

```python
file.describe()
```

```python
file['WSR0'].isnull().sum()
```

```python
file['WSR0'].fillna(0)
```

```python
file['WSR0'].fillna(file['WSR0'].mean(), inplace=True)
```

```python
file.set_index('Date', inplace=True)
```

```python
file['WSR0'].rolling(100).mean().plot();
```
